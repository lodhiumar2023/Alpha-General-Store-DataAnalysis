```python
import pandas as pd
```

```python
rewards_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/01_data_cleaning/RewardsData.csv')
rewards_df.head(20)
```

```python
# view missing values in the data frame
rewards_df.isna()
```

```python
# count missing values in data frame
missing_values_count = rewards_df.isna().sum()
missing_values_count
```

```python
# Display pecenatage of missing values
missing_values_percentages = missing_values_count / len(rewards_df) * 100
missing_values_percentages
```

```python
# Last Seen Values in Generic Object
rewards_df['Last Seen']
```

```python
# convert generic object to datetime format
rewards_df['Last Seen'] = pd.to_datetime(rewards_df['Last Seen'])
```

```python
rewards_df['Last Seen']
```

```python
rewards_df['Last Seen'].max()
```

```python
rewards_df['Last Seen'].min()
```

```python
# fill null values with last seen min
rewards_df['Last Seen'].fillna(value=rewards_df['Last Seen'].min(), inplace=True)
```
